```python
def downsample_img(img: np.array, sample_shape: (int, int)) -> np.array:
    """
    Downsamples the image img so that the new iamge has width = sample_width 
    and height = sample_height. The downsampled_img, is such that for each 
    coordinate (x, y) downsampled_img[x][y] is equal to the pixel value at
    img[x * img.height/sample_height][y * img.width/sample_width]
    """
    partition_height = int(np.ceil(img.shape[0] / sample_shape[0]))
    partition_width = int(np.ceil(img.shape[1] / sample_shape[1]))
    return img[::partition_height,::partition_width]
```

```python
def to_monochromatic(img):
    r_weight = 0.114
    g_weight = 0.299
    b_weight = 0.587
    return r_weight * img[:,:,0] + g_weight * img[:,:,1] + b_weight * img[:,:,2]
```

```python
def normalize_img(img, new_min=0, new_max=255):
    """ 
    Normalizes the image img to the range [new_min, new_max]. This is done by 
    applying to all pixels p a funciton f(p) = ap + b such that for img.max(),
    f(img.max()) = new_max and for img.min(), f(img.min()) = new_min.
    """
    img_min = img.min()
    img_max = img.max()
    a = (new_max - new_min) / (img_max - img_min)
    b = new_min - img_min * (new_max - new_min) / (img_max - img_min)
    return (a * img + b).astype(np.uint8)
```

```python
def threshold(img, percentile):
    return (img > np.percentile(img, percentile)).astype(int)
```

```python
def build_patches(img, patches_grid_order):
    rows = np.split(img, patches_grid_order)
    patches_grid = [np.split(img, patches_grid_order, axis=1) for img in rows]
    patches_list = reduce(operator.add, patches_grid, [])
    patches_h = patches_grid[0][0].shape[0]
    patches_w = patches_grid[0][0].shape[1]
    centers = []
    for i in range(len(patches_grid)):
        for j in range(len(patches_grid[0])):
            centers.append((i * patches_h + patches_h // 2, j * patches_w + patches_w // 2))
    centers.pop()
    patches_list.pop()
    return patches_list, centers
```

```python
def argsort_patches(patches, descriptor_fn, distance_fn, reference_patch):
    descriptors = [descriptor_fn(patch) for patch in patches]
    reference = descriptor_fn(reference_patch)
    distances = [distance_fn(descriptor, reference) for descriptor in descriptors]
    return np.argsort(distances)
```

```python
def euclidean_distance(x, y):
    return np.linalg.norm(x - y)
```

```python
def lbp_descriptor(patch, n_samples=8, radius=3):
    lbp = local_binary_pattern(patch, n_samples, radius, method='uniform')
    hist_bins = np.arange(0, n_samples + 3)
    hist_range = (0, n_samples + 2)
    histogram, _ = np.histogram(lbp, bins=hist_bins, range=hist_range)
    return histogram
```

```python
def argsort_patches_with_lbp(patches, template, n_samples=8, radius=3):
    descriptor = lambda patch: lbp_descriptor(patch, n_samples, radius)
    return argsort_patches(patches, descriptor, euclidean_distance, template)
```

```python
def estimate_center(img):
    image_white_area = np.sum(img)
    x_estimator = np.sum(np.arange(img.shape[0]) * img.T) / image_white_area
    y_estimator = np.sum(np.arange(img.shape[1]) * img) / image_white_area
    return (int(round(x_estimator)), int(round(y_estimator)))
```

```python
def image_center(patch_center_in_img, patch, pattern_center):
    patch_origin_x = patch_center_in_img[0] - patch.shape[0] // 2
    patch_origin_y = patch_center_in_img[1] - patch.shape[1] // 2
    pattern_center_x, pattern_center_y = pattern_center
    return np.array([patch_origin_x+pattern_center_x, patch_origin_y+pattern_center_y])
```

```python
def classify_coordinates(coordinates, correspondent_patches, img, template):
    bins = [[], [], [], []]
    x_middle = img.shape[0] // 2
    y_middle = img.shape[1] // 2
    for index, coordinate in enumerate(coordinates):
        if coordinate[0] > x_middle:
            if coordinate[1] > y_middle:
                bins[3].append(index)
            else:
                bins[2].append(index)
        else:
            if coordinate[1] > y_middle:
                bins[1].append(index)
            else:
                bins[0].append(index)
    for index, b in enumerate(bins):
        if len(b) > 1:
            bins[index] = [np.argmin([
                euclidean_distance(lbp_descriptor(correspondent_patches[i]),
                                   lbp_descriptor(template)) for i in b
            ])]
    return [coordinates[b[0]] if len(b) > 0 else [] for b in bins]
```

```python
def extract_main_grid_coordinates(preprocessed_img, template):
    patches, centers = build_patches(preprocessed_img, 8)
    sorted_patches_indices = argsort_patches_with_lbp(patches, template)
    template_matches = argsort_patches_with_lbp(patches, template)[:4]
    selected_patches = [patches[match] for match in template_matches]   
    selected_centers = [centers[match] for match in template_matches]
    estimated_patch_centers = [estimate_center(patch) for patch in selected_patches]
    estimated_image_centers = [image_center(center_img, patch, center_patch) 
                               for center_img, patch, center_patch 
                               in zip(selected_centers, selected_patches, estimated_patch_centers)]
    return classify_coordinates(estimated_image_centers, selected_patches, preprocessed_img, template)
```

```python
# Calculates the symmetric vertical padding that makes the 
# height of the smaller image img2 be equal to the one of img1.
# Returns (padding_up, padding_down).
def calculate_symmetric_vertical_padding(img1, img2):
    height_difference = img1.shape[0] - img2.shape[0]
    padding = height_difference // 2
    if height_difference % 2 != 0:
        return (padding, padding+1)
    return (padding, padding)
```

```python
def stitch_row(row_images, corners_coordinates):
    plane = row_images[0]
    for index in range(1, len(row_images)):
        prev_coordinates = corners_coordinates[index-1]
        current_coordinates = corners_coordinates[index]
        
        first_vote_participation = 0
        first_distance_x = 0
        first_distance_y = 0
        if len(prev_coordinates[1]) > 1 and len(current_coordinates[0]) > 1:
            first_distance_x = prev_coordinates[1][0] - current_coordinates[0][0]
            first_distance_y = plane.shape[1] + current_coordinates[0][1] - prev_coordinates[1][1]
            first_vote_participation = 1
            
        second_vote_participation = 0
        second_distance_x = 0
        second_distance_y = 0
        if len(prev_coordinates[3]) > 1 and len(current_coordinates[2]) > 1:
            second_distance_x = prev_coordinates[3][0] - current_coordinates[2][0]
            second_distance_y = plane.shape[1] + current_coordinates[2][1] - prev_coordinates[3][1]
            first_vote_participation = 0.5
            second_vote_participation = 0.5
            
        final_distance_x = int(first_vote_participation * first_distance_x\
            + second_vote_participation * second_distance_x)
        final_distance_y = int(first_vote_participation * first_distance_y\
            + second_vote_participation * second_distance_y)

        padding_right = row_images[index].shape[1] - final_distance_y
        y_stitch_line = plane.shape[1] - final_distance_y
        
        x_up_bound, x_low_bound = calculate_symmetric_vertical_padding(plane, row_images[index])
        assert row_images[index].shape[0] + x_up_bound + x_low_bound == plane.shape[0]
        x_up_bound += final_distance_x
        x_low_bound += -final_distance_x
        
        padding_up = 0
        padding_down = 0
        if x_low_bound < 0:
            padding_down = -x_low_bound
        if x_up_bound < 0:
            padding_up = -x_up_bound
        
        print((padding_up, padding_down))
        print((x_up_bound, x_low_bound))
        
        
        plane = np.pad(plane, ((padding_up, padding_down), (0, padding_right)),
                       mode='constant', constant_values=0)
        
        x_up_bound += padding_up
        x_low_bound += padding_down
        
        if x_low_bound == 0:
            plane[x_up_bound:,y_stitch_line:] = row_images[index]
        else:
            plane[x_up_bound:-x_low_bound,y_stitch_line:] = row_images[index]
    return plane
```
